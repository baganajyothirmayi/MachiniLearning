```python
import pandas as pd
```

```python
df=pd.read_table("fruit_data_with_colors.txt")
df.head(10)
```

```python
df.loc[df["fruit_label"]!=1,"fruit_label"]=0
df
```

```python
df["fruit_label"].value_counts()
```

```python
x=df[["mass","width","height"]]
x
```

```python
y=df["fruit_label"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logModel=LogisticRegression()
logModel.fit(x_train,y_train)
```

```python
x
```

```python
logModel.predict([[186,7.2,8.2],[192,8.4,7.3]])
```

```python
logModel.score(x_train,y_train)
```

```python
logModel.score(x_test,y_test)
```

## Example

```python
df=pd.read_table("fruit_data_with_colors.txt")
df.head(10)
```

```python
df["fruit_label"].value_counts()
```

```python
x1=df[["mass","width","height"]]
x1
```

```python
y1=df["fruit_label"]
y1
```

```python
from sklearn.model_selection import train_test_split
```

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logModel=LogisticRegression()
logModel.fit(x1_train,y1_train)
```

```python
logModel.predict([[186,7.2,8.2],[192,8.4,7.3]])
```

```python
logModel.score(x1_train,y1_train)
```

```python
logModel.score(x1_test,y1_test)
```
