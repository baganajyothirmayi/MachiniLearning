```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("Superstore.xls")
df
```

```python
fun1=df[["Order Date","Sales"]]
fun1=fun1.sort_values("Order Date")
fun1=fun1.groupby("Order Date")["Sales"].sum().reset_index()
fun1=fun1.set_index("Order Date")
fun1
```

```python
y=fun1["Sales"].resample("M").sum()
y
```

```python
y.plot()
```

```python
y.plot(kind="barh",figsize=[5,10])
```

```python
z=fun1["Sales"].resample("Y").sum()
z
```

```python
z.plot()
```
